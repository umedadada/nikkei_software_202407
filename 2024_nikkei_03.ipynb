```python
# ライブラリの読み込み
import pandas as pd #データを取り扱うためのライブラリ
import plotly.express as px #データを可視化するためのライブラリ
import plotly.graph_objs as go  #データを可視化するためのライブラリ
from plotly.graph_objs import Scatter, Figure, Layout, Histogram #データを可視化するためのライブラリ
from collections import Counter #度数表を作成するときに使用するライブラリ
```

```python
# データの読み込み
df=pd.read_csv('./baseball_player.csv')
# データのカラム確認
df.info()
# データの表示
df
```

```python
# 身長のリスト
heights=df['height']

# 5cmの区間にデータを変換する
heights=(height // 5 * 5 for height in heights)

# ヒストグラムの度数表を作成
height_counts = Counter(heights)
count_df = pd.DataFrame(height_counts.items(), columns=['区間(5cmごと)', '度数'])# データフレームに変換
count_df = count_df.sort_values('区間(5cmごと)')
count_df
```

```python
fig=px.bar(count_df,x='区間(5cmごと)',y='度数')
fig
```

```python
# ヒストグラム
fig = px.histogram(df, x="height", nbins=10,barmode='overlay')
fig.show()
```

```python
# ヒストグラム
fig = px.histogram(df, x="height", nbins=50,barmode='overlay')
fig.show()
```

```python
# ヒストグラム
fig = px.histogram(df, x="height", nbins=50, color="JPNorNot",barmode='overlay')
fig.show()
```

```python
# 散布図
fig = px.scatter(df, x="height", y="weight")
fig.show()
```

```python
# 散布図の色を指定
scatter_color = 'rgba(0, 0, 255, 0.7)'  # 例として青を指定

#Create Graph Object
scat = Scatter(
    name="Scatter",
    x = df['height'],
    y = df['weight'],
    mode = 'markers',
    marker=dict(color=scatter_color)  # 散布図の色を指定
)
hist_x = Histogram(
    name="Histogram_X",
    x = df['height'],
    yaxis='y2',
    marker=dict(color=scatter_color)  # 散布図の色をヒストグラムの色に設定
)
hist_y = Histogram(
    name="Histogram_Y",
    y = df['weight'],
    xaxis='x2',
    marker=dict(color=scatter_color)  # 散布図の色をヒストグラムの色に設定
)

data = [scat,hist_x,hist_y]

layout = Layout(
    title = "散布図と縦軸・横軸それぞれのヒストグラム",

    xaxis=dict(
        title="height",
        domain=[0, 0.85],
        showgrid=True
    ),
    yaxis=dict(
        title="weight",
        domain=[0, 0.85],
        showgrid=True
    ),
    xaxis2=dict(
        domain=[0.85, 1],
    ),
    yaxis2=dict(
        domain=[0.85, 1],
    )
)
fig = Figure(data=data, layout=layout)
fig
```

```python
# ヒストグラム
fig = px.histogram(df, x="salary", color="team",barmode='overlay')
fig.show()
```

```python
# 箱ひげ図
fig = px.box(df, x="team",y="salary", color = "team")
fig.show()
```

```python
# 箱ひげ図付きヒストグラム
# データをチームSに絞り込む
steam_df = df[df['team']=='S']
# 可視化
fig = px.histogram(steam_df, x="salary",color = "team",marginal='box')
fig
```
