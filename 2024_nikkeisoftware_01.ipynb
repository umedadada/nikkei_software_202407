```python
# ライブラリの読み込み
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
# データの読み込み
df=pd.read_csv('./FEH_00200521_240311192951.csv',encoding="shift-jis")
# データのカラム確認
df.info()

# データの表示
#行の表示数の上限を撤廃
pd.set_option('display.max_rows', None)
#列の表示数の上限を撤廃
pd.set_option('display.max_columns', None)
# データの表示
df
```

```python
# 可視化のためのデータ加工
## 条件の絞り込み
viz_df = df[(df['表章項目']=='人口') &
            (df['男女_時系列']=='総数') &
            (~df['地域_時系列'].isin(['全国', '人口集中地区', '人口集中地区以外の地区']))&
            (df['時間軸（調査年）']=='2020年')]
## 可視化する指標の型変換
viz_df = viz_df.assign(value = viz_df['value'].astype('int64'))
## 軸の名前を明示的に指定して可視化
fig=px.bar(viz_df, x='地域_時系列', y='value')
fig.update_layout(
    title='図1. 2020年の各都道府県の人口',
    xaxis=dict(title='都道府県'),
    yaxis=dict(title='人口'))
```

```python
## ソート
sorted_viz_df = viz_df.sort_values(by='value',ascending=False)
## 軸の名前を明示的に指定して可視化
fig=px.bar(sorted_viz_df, x='地域_時系列', y='value')
fig.update_layout(
    title='図2. 2020年の各都道府県の人口',
    xaxis=dict(title='都道府県'),
    yaxis=dict(title='人口'))
```
